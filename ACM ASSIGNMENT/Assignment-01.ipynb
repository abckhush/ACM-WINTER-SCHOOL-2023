```python
import pandas as pd
```

```python
file_path = 'C:\\Users\\Khushi\\Downloads\\Advertising.csv'
```

```python
advertising_data = pd.read_csv(file_path)
advertising_data
```

```python
advertising_data.info()
```

```python
advertising_data.describe()
```

```python
print (advertising_data.columns)
```

```python
# Visualizing the data
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of TV advertising vs. sales
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TV', y='sales', data=advertising_data)
plt.title('TV Advertising vs. Sales')
plt.xlabel('TV Advertising ($)')
plt.ylabel('Sales (units)')
plt.show()
```

```python
sns.jointplot(x='TV', y='sales', data=advertising_data, kind='scatter')
sns.jointplot(x='radio', y='sales', data=advertising_data, kind='scatter')
sns.jointplot(x='newspaper', y='sales', data=advertising_data, kind='scatter')
```

```python
sns.heatmap(advertising_data.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
```

```python
X = advertising_data[['TV', 'radio', 'newspaper']] 
y = advertising_data['sales']
```

```python
# Spliting the data in Training and Testing Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Choosing a Regression Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Evaluation
from sklearn.metrics import mean_squared_error, r2_score
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r_squared = r2_score(y_test, y_pred)

print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared: {r_squared}')
```

```python
# Predictions
new_X = [[100, 20, 30]]
predicted_sales = model.predict(new_X)

print(f'Predicted sales: {predicted_sales[0]}')
```

```python
# CREDIT.csv DATASET
```

```python
file_path1= 'C:\\Users\\Khushi\\Downloads\\Credit.csv'
```

```python
credit_data = pd.read_csv(file_path1)
credit_data
```

```python
credit_data.info()
```

```python
credit_data.describe()
```

```python
print (credit_data.columns)
```

```python
# Converting the strings into numerical values: One-hot encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
credit_data['Married'] = label_encoder.fit_transform(credit_data['Married'])
credit_data['Gender'] = label_encoder.fit_transform(credit_data['Gender'])
credit_data['Student'] = label_encoder.fit_transform(credit_data['Student'])
credit_data['Ethnicity'] = label_encoder.fit_transform(credit_data['Ethnicity'])
print (credit_data)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X = credit_data.drop('Defaultee', axis=1)
y = credit_data['Defaultee']
```

```python
# Spliting the dataset into Training and Testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Training
model = LogisticRegression(max_iter=10000)
model.fit(X_train, y_train)
```

```python
# Predicting
y_pred = model.predict(X_test)
```

```python
print('Accuracy:', accuracy_score(y_test, y_pred))
print('\nClassification Report:')
print(classification_report(y_test, y_pred))
print('\nConfusion Matrix:')
print(confusion_matrix(y_test, y_pred))
```

```python
# Using SVM
```

```python
from sklearn.svm import SVC

model = SVC()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
print('Accuracy:', accuracy_score(y_test, y_pred))
print('\nClassification Report:')
print(classification_report(y_test, y_pred))
print('\nConfusion Matrix:')
print(confusion_matrix(y_test, y_pred))
```
