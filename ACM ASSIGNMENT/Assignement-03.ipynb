```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical,plot_model
```

```python
dataset= pd.read_csv('/content/drive/MyDrive/data.csv')
dataset
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
# Split the dataset into training and testing sets
columns_to_drop = ['diagnosis', 'Unnamed: 32']
dataset_dropped = dataset.drop(columns=columns_to_drop)

x_train, x_test, y_train, y_test = train_test_split(dataset_dropped, dataset['diagnosis'], test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape
```

```python
# Scaling the features
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test= scaler.transform(x_test)
```

```python
# Label Encoding
label=LabelEncoder()
y_train= label.fit_transform(y_train)
y_test=label.transform(y_test)
```

```python
# Building the ANN Model
model= Sequential([
    Dense(64, activation='relu', input_shape=(x_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(3, activation='softmax')
])
```

```python
model.summary()
```

```python
# Ploting the ANN Model
plot_model(model, to_file='model.png', show_shapes=True, show_layer_names=True)
plt.figure(figsize=(8,6))
img= plt.imread('model.png')
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# Compiling the model
model.compile(optimizer='adam', loss='mean_squared_error')
# Training the model and saving its history
h=model.fit(x_train, y_train, epochs=100, batch_size=32, validation_split=0.1)
```

```python
# Evaluate the model on the test set
test_loss=model.evaluate(x_test,y_test)
print('Test Loss:', test_loss)
```

```python
plt.plot(h.history['loss'],label='Training Loss')
plt.plot(h.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
